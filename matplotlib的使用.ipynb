```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3,4,6,1,10,7,5,2])
plt.plot(x, y, 'r') #绘制折线图 'r'表示红色
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3,4,6,1,10,7,5,2])
plt.plot(x, y, 'b', lw=5) # lw = line width 线段粗细
#柱状图
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3,5,1,8,12,0.5,4,7])
plt.bar(x, y, 0.2, alpha=0.2, color='r') # 第三个参数表示柱状占比，1的话就是占满一个单元
plt.show()
```
